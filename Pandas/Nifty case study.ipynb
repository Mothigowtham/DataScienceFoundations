## **Task**

1. <font color = 'Green'>**What fraction of the days did the markets close higher than the previous days.**

## **Problem 1**

```python
len(nifty)
```

```python
nifty.values[1:] - nifty.values[:-1]
```

```python
np.sum(nifty.values[1:] - nifty.values[:-1] > 0)
```

```python
np.sum(nifty.values[1:] - nifty.values[:-1] > 0)/len(nifty)
```

***Here, we could conclude that more than half of the year markets close is higher than the previous days close.***

## **Problem 3**

```python
dow = new_nifty.copy()
```

```python
for i in dow.index:
    dow[i] = i.dayofweek
```

```python
friday_subset = new_nifty[dow == 4]
```

```python
friday_subset
```

***We can now able to see the subset containing the market values of all fridays in the year.***
