### **Creating Data Frame objects using function**

```python
def df(nRows, mCols, maxRand):
    arr = np.random.randint(0, maxRand, (nRows, mCols))
    df = pd.DataFrame(arr)
    df.index = ['R' + str(x) for x in np.arange(1, nRows + 1)]
    df.columns = ['C' + str(x) for x in np.arange(1, mCols + 1)]
    return df
```

# Dropping Rows with null values


<font color = 'red'> ***Going through each row of the dataframe and deleting if any of the columns is null***

## Method 1: conventional way of dropping row if it has null values

```python
for r in planets.index:
    for c in planets.columns:
        if pd.isnull(planets.loc[r, c]):
            planets.drop(r, inplace = True)
            break
```

```python
planets.describe
```

##Method 2: Faster way of dropping row having null values



```python
planets = sns.load_dataset('planets')
```

```python
planets.info()
```

```python
for i, r in planets.iterrows():
    if pd.isnull(r).any():
        planets.drop(i, inplace = True)
```

```python
planets.info()
```

## Method 3: Ultimate faster way of droppinng row having null values

```python
planets = sns.load_dataset('planets')
```

```python
planets.info()
```

```python
planets.dropna(inplace = True)
```

```python
planets.info()
```

 # **Filtering Dataset**

Filter and show only those rows which have planets that are found in the 2010's and method is 'Radial Velocity' and 'Transit' and distance is large (>=75 percentile)


```python
planets_copy = planets.copy()
```

```python
percentile_75 = planets['distance'].quantile(0.75)
```

```python
percentile_75
```

```python
 for i, r in planets_copy.iterrows():
     if r['year'] < 2010:
         planets_copy.drop(i, inplace = True)
         continue
     if r['method'] != 'Radial Velocity' and r['method'] != 'Transit':
         planets_copy.drop(i, inplace = True)
         continue
     if r['distance'] < percentile_75:
         planets_copy.drop(i, inplace = True)
         continue
```

```python
planets_copy
```

```python
planets_copy.info()
```

```python
df = sns.load_dataset('planets')
```

```python
df.info()
```

```python
df = df.copy()
df = df[
        (df['year'] >= 2010) &
        ((df['method'] == 'Transit') | (df['method'] == 'Radial Velocity')) &
        (df['distance'] > percentile_75)
        ]
```
