### Q12. What is the maximum duration of Movie/TV Show in Netflix?

```python
movie_df = netflix_df[netflix_df['type'] == 'Movie']
```

```python
movie_df['run_time'] = movie_df['duration'].apply(lambda x: x.split(' ')[0])
```

```python
movie_df['run_time'] = movie_df['run_time'].astype('int')
```

```python
top5_movies = movie_df[['title', 'run_time']].sort_values('run_time', ascending = False).head(5)
top5_movies
```

```python
plt.figure(figsize = (10,8))

sns.barplot(x = top5_movies['title'], y = top5_movies['run_time'])
plt.xlabel('Movies', fontsize = 15)
plt.ylabel('Run Time', fontsize = 15)
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title('Top 5 Movie with highest Runtime', fontsize = 20, fontweight = 'bold')

plt.show()
```

```python
show_df = netflix_df[netflix_df['type'] == 'TV Show']

show_df['run_time'] = show_df['duration'].apply(lambda x: x.split(' ')[0])

show_df['run_time'] = show_df['run_time'].astype('int')
```

```python
top5_shows = show_df[['title', 'run_time']].sort_values('run_time', ascending = False).head(5)
top5_shows
```

```python
plt.figure(figsize = (10,8))

sns.barplot(x = top5_shows['title'], y = top5_shows['run_time'])
plt.xlabel('TV Shows', fontsize = 15)
plt.ylabel('Run Time', fontsize = 15)
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title('Top 5 TV Shows with highest Runtime', fontsize = 20, fontweight = 'bold')

plt.show()
```
