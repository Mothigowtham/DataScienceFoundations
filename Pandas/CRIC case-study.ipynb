```python
def stats(col):
    print('mean: ', np.mean(col))
    print('median: ', np.median(col))
    print('IQR: ', np.percentile(col, 75) - np.percentile(col, 25))
```

## **Problem 6**
<font color = 'Green'>**Find mean for Sachin's scores based on which quartile India's score falls** 


```python
india_quartile = np.percentile(india, [25,50,75,100])
```

```python
india_quartile = india_quartile.reshape(4,1)
```

```python
indices = india < india_quartile
```

```python
sachin[indices[1, :]]
```

```python
for i in range(4):
    print(i, np.mean(sachin[indices[i]]))
```
